# Activity 3 - K-means clustering with Pew Data

#### Reading data from CSV

```r
pew <- read.csv("June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv")
```

#### Extracting required data and cleaning it

```r
data <- data.frame(Age = pew$age, Date=pew$date1a, Game = pew$game1, Empl = pew$emplnw, Education = pew$educ2)
clean_data <- na.omit(data)
clean_data <- clean_data[!clean_data$Date %in% c(8,9, NA),]
clean_data <- clean_data[!clean_data$Game %in% c(8,9, NA),]
clean_data <- clean_data[!clean_data$Empl %in% c(98,99, NA),]
clean_data <- clean_data[!clean_data$Age %in% c(98,99, NA),]
clean_data <- clean_data[!clean_data$Education %in% c(98,99, NA),]
head(clean_data)
```

#### Scaling the cleaned data

```r
scaled_data<- data.frame(scale(clean_data))
scaled_data
```

### Clustering data for k = 5

```r
set.seed(76964057) #Set the seed for reproducibility
k1 <-kmeans(scaled_data, centers=5) 
k1$centers
table(k1$cluster)
```
